## Univariate Data Analysis.

### Non-Graphical.

```python
for col in dataFrame.columns:
  print(col, "Information : ")
  print("Mean : ", dataFrame[col].mean())
  print("Median : ", dataFrame[col].median())
  print("Mode : ", dataFrame[col].mode())

  print("Minimum : ", dataFrame[col].min())
  print("Maximum : ", dataFrame[col].max())

  print("Quantiles : ", numpy.quantile(dataFrame[col], [0, 0.25, 0.5, 0.75, 1]))

  print("Variance : ", numpy.var(dataFrame[col], ddof=1))
  print("Standard Deviation : ", numpy.std(dataFrame[col], ddof=1))

  print("Skew : ", dataFrame[col].skew())
  print("Kurtosis : ", dataFrame[col].kurtosis())
  print()
```
