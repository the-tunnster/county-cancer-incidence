```python
df_train = pandas.read_csv(base_file_path + "training.csv")
df_test = pandas.read_csv(base_file_path + "testing.csv")
```

# Checking for outlier values at a high level.

```python
for col in dataFrame.columns:
  if col == "binnedInc" or col == "Geography":
    continue
  print(col, ": ")

  print("\tMean : ", dataFrame[col].mean())
  print("\tMedian : ", dataFrame[col].median())

  print("\tMinimum : ", dataFrame[col].min())
  print("\tMaximum : ", dataFrame[col].max())

  print()
```

```python
for i in range (dataFrame.shape[0]):
  if (dataFrame["MedianAge"][i] > 100):
    dataFrame["MedianAge"][i] = (dataFrame["MedianAgeMale"][i] + dataFrame["MedianAgeFemale"][i])/2

print(dataFrame["MedianAge"].max())
```

# Checking for NaN values.

```python
dataFrame[dataFrame.isna().any(axis=1)]
```

```python
for col in dataFrame.columns:
  nan_counts = dataFrame[col].isna().sum()
  if nan_counts > 0:
    print(col, ":", nan_counts)
```

```python
dataFrame.drop(columns=["Geography", "binnedInc", "PctSomeCol18_24", "PctPrivateCoverageAlone", "Id"], inplace=True)
```

```python
for col in dataFrame.columns:
  nan_counts = dataFrame[col].isna().sum()
  if nan_counts > 0:
    print(col, ":", nan_counts)
```

```python
df_cleaned = dataFrame.dropna(how="any")
df_cleaned.shape
```

```python
df_cleaned.to_csv(base_file_path + "cleaned.csv", index=False)
```

# Creating a normalised dataset.

```python
df_normalised = dataFrame.dropna(how="any")
df_normalised.shape
```

```python
columns_to_normalise = ["avgAnnCount", "popEst2015", "studyPerCap", "medIncome"]
normalisedFrame = df_normalised[columns_to_normalise].copy()
```

```python
df_normalised[columns_to_normalise] = (normalisedFrame-normalisedFrame.mean())/normalisedFrame.std()
df_normalised.head()
```

```python
df_normalised.to_csv(base_file_path + "normalised.csv", index=False)
```
